## (b) iii.

```python
test_X = pd.concat([class0_df.iloc[0:200,0:4],class1_df.iloc[0:200,0:4]]).reset_index(drop=True)
test_y = pd.concat([class0_df.iloc[0:200,[4]],class1_df.iloc[0:200,[4]]]).reset_index(drop=True)
train_X = pd.concat([class0_df.iloc[200:,0:4],class1_df.iloc[200:,0:4]]).reset_index(drop=True)
train_y = pd.concat([class0_df.iloc[200:,[4]],class1_df.iloc[200:,[4]]]).reset_index(drop=True)
print (len(class0_df))
print (len(class1_df))
```

## (c) i & ii
I used smaller increments = 1 of k here, 

```python
# from imblearn.over_sampling import SMOTE
# sm = SMOTE(random_state=42)
# print (len(train_y))
# train_X, train_y = sm.fit_sample(train_X, train_y)
# print (len(train_y))
```

```python

from sklearn.neighbors import KNeighborsClassifier
train_error=[]
test_error=[]
# here k = i:
for i in range(1,902):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(train_X, train_y)
    train_error.append(1-neigh.score(train_X, train_y))
    test_error.append(1-neigh.score(test_X, test_y))
    print("n =",i)
    print("train error:")
    print (1-neigh.score(train_X, train_y))
    print("test error:")
    print (1-neigh.score(test_X, test_y))   
```

#### plot

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20,10))
plt.plot([1/x for x in range(1,902)],train_error,'^',label='train error')
plt.plot([1/x for x in range(1,902)],test_error,'*',label='test error')
pylab.legend(loc='upper right')
plt.xlabel('1/k', fontsize=18)
plt.ylabel('error rate', fontsize=16)
plt.show()
```

```python
#  we can get the result that when n<=18 the train and test error both are 0.
#  So I need to calculate the confusion matrix, true positive rate, 
#   true negative rate, precision, and F-scoreF-score when k <=18.
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
neigh = KNeighborsClassifier(n_neighbors=18)
neigh.fit(train_X, train_y)
y_pred = neigh.predict(test_X)
y_true = test_y
print ("confusion matrix:")
print (confusion_matrix(y_true,y_pred))
print ("true positive rate:")
print (confusion_matrix(y_true,y_pred)[0][0]/(len(test_y)/2))
print ("true negative rate:")
print (confusion_matrix(y_true,y_pred)[1][1]/(len(test_y)/2))
print ("precision:")
print (neigh.score(test_X, test_y))
print ("F-score:")
print (f1_score(y_true,y_pred))
```

## (c) iii Learning Curve.

```python
train_error_plot = []
test_error_plot = []
train_bestk_plot = []
test_bestk_plot = []
for N in range(50,801,50):
    sub_train_error=[]
    sub_test_error=[]
    subtrain_X = pd.concat([class0_df.iloc[200:201+N//2,0:4],class1_df.iloc[200:201+N//2,0:4]]).reset_index(drop=True)
    subtrain_Y = pd.concat([class0_df.iloc[200:201+N//2,[4]],class1_df.iloc[200:201+N//2,[4]]]).reset_index(drop=True)
    for k in range(1,N+1,40):
        neigh = KNeighborsClassifier(n_neighbors=k)
        neigh.fit(subtrain_X, subtrain_Y)
        sub_train_error.append(1-neigh.score(subtrain_X, subtrain_Y))
        sub_test_error.append(1-neigh.score(test_X, test_y))
    train_error_plot.append(min(sub_train_error))
    train_bestk_plot.append(sub_train_error.index(min(sub_train_error))*40+1)
    test_error_plot.append(min(sub_test_error))
    test_bestk_plot.append(sub_test_error.index(min(sub_test_error))*40+1)
```

```python
print("best k for every size for train error")
print(train_bestk_plot)
print("best k for every size for test error")
print(test_bestk_plot)
sns.set_style("darkgrid")
plt.figure(figsize=(20,10))
plt.plot(range(50,801,50),train_error_plot,'^',label='train error')
plt.plot(range(50,801,50),test_error_plot,'*',label='test error')
pylab.legend(loc='upper right')
plt.xlabel('train dataset size', fontsize=18)
plt.ylabel('best error rate', fontsize=16)
plt.show()
```

### B

```python
#  I used k=1 here.
import numpy as np
test_error = []
for p in np.logspace(0.1,1, num=10, base=10):
    neigh = KNeighborsClassifier(n_neighbors=1,metric='minkowski', metric_params={'p': p})
    neigh.fit(train_X, train_y)
    test_error.append(1-neigh.score(test_X, test_y))
    print("p =",p)
    print("test error:")
    print (1-neigh.score(test_X, test_y))  
# the best p is 10 with error = 0
```

### c which becomes Chebyshev Distance with p → ∞

```python
test_error = []
for i in range(1,902,10):
    neigh = KNeighborsClassifier(n_neighbors=i,metric='chebyshev')
    neigh.fit(train_X, train_y)
    test_error.append(1-neigh.score(test_X, test_y))
    print("n =",i)
    print("test error:")
    print (1-neigh.score(test_X, test_y)) 
best_k_manhattan = test_error.index(min(test_error))*10+1
print (best_k_manhattan)
# the best k is still 1 with error = 0.
```

## (d) ii Mahalanobis Distance.

```python
from sklearn.neighbors import DistanceMetric
test_error = []
for i in range(1,902,10):
    neigh = KNeighborsClassifier(n_neighbors=i, metric='mahalanobis', metric_params={'VI': np.linalg.pinv(np.cov(np.array(train_X)))}, algorithm="brute")
    neigh.fit(train_X, train_y)
    test_error.append(1-neigh.score(test_X, test_y))
    print("n =",i)
    print("test error:")
    print (1-neigh.score(test_X, test_y)) 
best_k_manhattan = test_error.index(min(test_error))*10+1
print (best_k_manhattan)
```
